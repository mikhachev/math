Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) 
и сравните результаты.


```python
j, num = 0, 10000
a = np.random.randint(0, 2, num)
b = np.random.randint(0, 2, num)
c = np.random.randint(0, 2, num)
d = np.random.randint(0, 2, num)

x = a + b + c + d
#print(a, b, c, x)
for i in range(0, num):
    if x[i] == 2:
        j = j + 1
print(j, num, j/num)
def binom(n, k):
    y = 2**(k-n)*2**(-k)*np.math.factorial(n)/(np.math.factorial(k)*np.math.factorial(n-k))
    return y
    

print('Вероятность для k=2, n = 4: ', binom(4, 2))

# Предсказанная вероятность по формуле соблюдается на практике
```

 Дополните код расчетом коэффициента корреляции x и y по

```python
n = 100
r = 0.9
x = np.random.rand(n)

y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.xlabel('y')
plt.grid(True)
plt.show()
print('Коэффициенты корреляции, встроенная формула и вручную')
c = np.corrcoef(x, y)
print(c)
xm = np.sum(x)/n
ym = np.sum(y)/n
R = np.sum((x-xm)*(y-ym))/np.sqrt(np.sum((x-xm)**2)*np.sum((y-ym)**2))
print( R)
```
