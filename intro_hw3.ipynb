Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
x = []
y1 = []
y2 =[]
y3 =[]
y4 =[]
y5 =[]
y6 =[]
x0 = 1
y0 = 1
R = 3
a = 2
b = 5
for i in range(-500, 500):
    x_ = i/100
    x.append(i/100)
    y1.append(np.sqrt(R**2 - (x_ - x0)**2) + y0)
    y2.append(-(np.sqrt(R**2 - (x_ - x0)**2)) + y0)
    y3.append(b*np.sqrt(1-x_**2/a**2))
    y4.append(-b*np.sqrt(1-x_**2/a**2))
    y5.append(b*np.sqrt(x_**2/a**2-1))
    y6.append(-b*np.sqrt(x_**2/a**2-1))
    
plt.title('График окружности')    
plt.plot(x,y1 , '-b')  
plt.plot(x,y2, '-b') 
```

```python
plt.title('График эллипса')    
plt.plot( x , y3 , '-b')  
plt.plot( x , y4 , '-b') 
```

```python
plt.title('График гиперболы')    
plt.plot( x , y5 , '-b')  
plt.plot( x , y6 , '-b') 
```

Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
x = []
y1 = []
y2 =[]

for i in range(-500, 500):
    x_ = i/100
    x.append(i/100)
    y1.append(x_**2 - 1)
    y2.append(1 + (np.exp(x_) -1)/x_)
    
   
plt.plot(x,y1 )  
plt.plot(x,y2)
```

```python
def equations(p):
    x, y = p
    return(y - x**2 + 1, np.exp(x) + x - x*y - 1)
x1, y1 = fsolve(equations, (-2, 2))
x2, y2 = fsolve(equations, (3, 6))
x3, y3 = fsolve(equations, (4, 20))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```
