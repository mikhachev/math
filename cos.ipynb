```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
x = np.linspace(0, 10, 151)
y1 = np.cos(x)
y2 = np.cos(4*x)
plt.axis([0.5, 9.5, -1.5, 1.5])
plt.plot(x, y1, color='green')
plt.plot(x, y2, color='purple')
plt.title('График f(t)', fontsize=16, fontweight='bold', color='DarkRed', loc='left')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```
